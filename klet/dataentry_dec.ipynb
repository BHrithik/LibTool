```python
@sync_to_async
def add_records():
    for _, row in df.iterrows():
        # Create a new Record for each row in the Excel file
        # Map the Excel columns to your Django model fields
        Record.objects.create(
            genre=row['Genre'],  # Assuming 'Genre' is the column name in your Excel file
            authorKorean=row['Author'],  # Assuming 'Author' is the column name for authorKorean in your Excel file
            authorEnglish=row['Author(Eng)'],  # and so on...
            workTitle=row['Translation Title'],
            workTitleKorean=row['Korean Title'],
            translator=row['Translator'],
            sourceTitle=row['Source'],
            publisher=row['Publisher'],
            year=row['Year'],
            # Add other fields as necessary
        )
```

```python
await add_records()
```
